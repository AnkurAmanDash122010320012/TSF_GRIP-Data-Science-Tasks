### **Step 1: Importing the necessary libraries and loading the dataset**

Loading the dataset and exploring its structure. This step involves importing the necessary libraries, loading the "SampleSuperstore" dataset, and examining its structure using the head() and info() functions.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the dataset
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/SampleSuperstore.csv")
```

### **Step 2: Data Exploration**

Checking for missing values. In this step, we check if the dataset contains any missing values using the isnull() function and visualize the missing values using a heatmap.


```python
# Displaying the first few rows of the dataset
df.head()
```

## **Step 3: Exploratory Data Analysis**




**a) Analysis using Pairplot of each column**

```python
sns.pairplot(df)
plt.show()
```

The pairplot displays scatter plots of all possible combinations of numerical columns in the dataset. It helps us visualize relationships between variables such as sales, profit, and discount.

**b) Based on Category**

```python
sns.pairplot(df, hue='Category')
plt.show()
```

This pairplot is similar to the previous one, but it adds color differentiation based on the 'Category' column. It allows us to observe the relationships between variables within each category separately.

**c) Based on Region**

```python
sns.pairplot(df, hue='Region')
plt.show()
```

This pairplot introduces color differentiation based on the 'Region' column. It helps us understand the relationships between variables within different regions.

**d) Based on Segment**

```python
sns.pairplot(df, hue='Segment')
plt.show()
```

Similarly, this pairplot introduces color differentiation based on the 'Segment' column. It helps us analyze the relationships between variables for different customer segments.

### **Step 4: Correlation Heatmap**

```python
correlation = df.corr()
sns.heatmap(correlation, annot=True)
plt.show()
```

The heatmap provides a visual representation of the correlation between different numerical variables. It helps us identify strong positive or negative correlations, which can provide insights into the factors affecting profit, sales, and discounts.

### **Step 5: Count plot of each column**

```python
fig, axes = plt.subplots(4, 2, figsize=(12, 16))

# Plot count plots
sns.countplot(ax=axes[0, 0], data=df, x='Category')
axes[0, 0].set_title('Count Plot - Category')

sns.countplot(ax=axes[0, 1], data=df, x='Region')
axes[0, 1].set_title('Count Plot - Region')

sns.countplot(ax=axes[1, 0], data=df, x='Segment')
axes[1, 0].set_title('Count Plot - Segment')

sns.countplot(ax=axes[1, 1], data=df, x='Ship Mode')
axes[1, 1].set_title('Count Plot - Ship Mode')

sns.countplot(ax=axes[2, 0], data=df, x='State')
axes[2, 0].set_title('Count Plot - State')
axes[2, 0].tick_params(axis='x', rotation=90)

sns.countplot(ax=axes[2, 1], data=df, x='City')
axes[2, 1].set_title('Count Plot - City')
axes[2, 1].tick_params(axis='x', rotation=90)

sns.countplot(ax=axes[3, 0], data=df, x='Sub-Category')
axes[3, 0].set_title('Count Plot - Sub-Category')
axes[3, 0].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

This code generates count plots for each categorical column in the dataset. It helps us understand the distribution and frequency of different categories, regions, segments, ship modes, states, cities, and sub-categories.

### **Step 6: Distribution of the data using plots**

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot distribution plots
sns.histplot(ax=axes[0, 0], data=df, x='Sales', bins=30, kde=True)
axes[0, 0].set_title('Distribution Plot - Sales')

sns.histplot(ax=axes[0, 1], data=df, x='Profit', bins=30, kde=True)
axes[0, 1].set_title('Distribution Plot - Profit')

sns.histplot(ax=axes[1, 0], data=df, x='Discount', bins=30, kde=True)
axes[1, 0].set_title('Distribution Plot - Discount')

sns.histplot(ax=axes[1, 1], data=df, x='Quantity', bins=30, kde=True)
axes[1, 1].set_title('Distribution Plot - Quantity')

plt.tight_layout()
plt.show()
```

This code generates distribution plots for numerical columns. It helps us visualize the distribution of sales, profit, discount, and quantity, allowing us to identify any skewness or outliers.

### **Step 7: Country-wise Deal Analysis**

```python
country_deals = df.groupby('Country')['Profit'].sum().reset_index().sort_values('Profit', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(data=country_deals, x='Country', y='Profit', palette='pastel')
plt.title('Country-wise Deal Analysis - Total Profit')
plt.xlabel('Country')
plt.ylabel('Total Profit')
plt.xticks(rotation=90)
plt.show()
```

This code calculates the total profit for each country and visualizes it using a bar plot. It helps us identify the countries with the highest and lowest profits.

### **Step 8: State-wise Deal Analysis**

```python
state_deals = df.groupby('State')['Profit'].sum().reset_index().sort_values('Profit', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(data=state_deals, x='State', y='Profit', palette='bright')
plt.title('State-wise Deal Analysis - Total Profit')
plt.xlabel('State')
plt.ylabel('Total Profit')
plt.xticks(rotation=90)
plt.show()
```

This code calculates the total profit for each state and visualizes it using a bar plot. It helps us identify the states with the highest and lowest profits.

### **Step 9: City-wise Analysis of Dealing**

```python
city_deals = df.groupby('City')['Profit'].sum().reset_index().sort_values('Profit', ascending=False).head(20)

plt.figure(figsize=(12, 6))
sns.barplot(data=city_deals, x='City', y='Profit', palette='bright')
plt.title('City-wise Deal Analysis - Top 20 Cities')
plt.xlabel('City')
plt.ylabel('Total Profit')
plt.xticks(rotation=90)
plt.show()
```

This code calculates the total profit for each city and visualizes the top 20 cities with the highest profits using a bar plot.

### **Step 10: Segment-wise Analysis of Profit, Discount, and Sales**

```python
segment_analysis = df.groupby('Segment')[['Profit', 'Discount', 'Sales']].sum().reset_index()

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot segment-wise profit
sns.barplot(ax=axes[0], data=segment_analysis, x='Segment', y='Profit', palette='Oranges_r')
axes[0].set_title('Segment-wise Profit')
axes[0].set_xlabel('Segment')
axes[0].set_ylabel('Total Profit')

# Plot segment-wise discount
sns.barplot(ax=axes[1], data=segment_analysis, x='Segment', y='Discount', palette='Reds_r')
axes[1].set_title('Segment-wise Discount')
axes[1].set_xlabel('Segment')
axes[1].set_ylabel('Total Discount')

# Plot segment-wise sales
sns.barplot(ax=axes[2], data=segment_analysis, x='Segment', y='Sales', palette='pastel')
axes[2].set_title('Segment-wise Sales')
axes[2].set_xlabel('Segment')
axes[2].set_ylabel('Total Sales')

plt.tight_layout()
plt.show()
```

This code calculates the total profit, discount, and sales for each customer segment (Consumer, Corporate, and Home Office). It visualizes the segment-wise analysis using three bar plots.

### **Step 11: Country-wise Analysis of Profit, Discount, and Sales**

```python
country_analysis = df.groupby('Country')[['Profit', 'Discount', 'Sales']].sum().reset_index()

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot country-wise profit
sns.barplot(ax=axes[0], data=country_analysis, x='Country', y='Profit', palette='Blues_r')
axes[0].set_title('Country-wise Profit')
axes[0].set_xlabel('Country')
axes[0].set_ylabel('Total Profit')
axes[0].tick_params(axis='x', rotation=90)

# Plot country-wise discount
sns.barplot(ax=axes[1], data=country_analysis, x='Country', y='Discount', palette='Blues_r')
axes[1].set_title('Country-wise Discount')
axes[1].set_xlabel('Country')
axes[1].set_ylabel('Total Discount')
axes[1].tick_params(axis='x', rotation=90)

# Plot country-wise sales
sns.barplot(ax=axes[2], data=country_analysis, x='Country', y='Sales', palette='Blues_r')
axes[2].set_title('Country-wise Sales')
axes[2].set_xlabel('Country')
axes[2].set_ylabel('Total Sales')
axes[2].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

This code calculates the total profit, discount, and sales for each country. It visualizes the country-wise analysis using three bar plots.

### **Step 12: State-wise Analysis of Profit, Discount, and Sales**

```python
state_profit = state_analysis['Profit'].abs()
state_discount = state_analysis['Discount'].abs()
state_sales = state_analysis['Sales']

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot state-wise profit
axes[0].pie(state_profit, labels=state_analysis['State'], autopct='%1.1f%%', colors=sns.color_palette('Reds_r'))
axes[0].set_title('State-wise Profit')

# Plot state-wise discount
axes[1].pie(state_discount, labels=state_analysis['State'], autopct='%1.1f%%', colors=sns.color_palette('deep'))
axes[1].set_title('State-wise Discount')

# Plot state-wise sales
axes[2].pie(state_sales, labels=state_analysis['State'], autopct='%1.1f%%', colors=sns.color_palette('pastel'))
axes[2].set_title('State-wise Sales')

plt.tight_layout()
plt.show()
```

This code calculates the total profit, discount, and sales for each state. It sorts the states based on profit in descending order and visualizes the state-wise analysis using three pie plots.

### **Step 13: City-wise Analysis of Profit, Discount, and Sales**

```python
city_analysis = df.groupby('City')[['Profit', 'Discount', 'Sales']].sum().reset_index().sort_values('Profit', ascending=False).head(20)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot city-wise profit
city_profit = city_analysis['Profit']
axes[0].pie(city_profit, labels=city_analysis['City'], autopct='%1.1f%%', colors=sns.color_palette('deep'))
axes[0].set_title('City-wise Profit - Top 20 Cities')

# Plot city-wise discount
city_discount = city_analysis['Discount']
axes[1].pie(city_discount, labels=city_analysis['City'], autopct='%1.1f%%', colors=sns.color_palette('dark'))
axes[1].set_title('City-wise Discount - Top 20 Cities')

# Plot city-wise sales
city_sales = city_analysis['Sales']
axes[2].pie(city_sales, labels=city_analysis['City'], autopct='%1.1f%%', colors=sns.color_palette('pastel'))
axes[2].set_title('City-wise Sales - Top 20 Cities')

plt.tight_layout()
plt.show()
```

This code calculates the total profit, discount, and sales for each city. It selects the top 20 cities based on profit in descending order and visualizes the city-wise analysis using three pie plots.

### **Step 14: Quantity-wise Sales, Profit, and Discount Analysis**

```python
quantity_analysis = df.groupby('Quantity')[['Sales', 'Profit', 'Discount']].sum().reset_index()

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot quantity-wise sales
quantity_sales = quantity_analysis['Sales']
axes[0].pie(quantity_sales, labels=quantity_analysis['Quantity'], autopct='%1.1f%%', colors=sns.color_palette('deep'))
axes[0].set_title('Quantity-wise Sales')

# Plot quantity-wise profit
quantity_profit = quantity_analysis['Profit']
axes[1].pie(quantity_profit, labels=quantity_analysis['Quantity'], autopct='%1.1f%%', colors=sns.color_palette('bright'))
axes[1].set_title('Quantity-wise Profit')

# Plot quantity-wise discount
quantity_discount = quantity_analysis['Discount']
axes[2].pie(quantity_discount, labels=quantity_analysis['Quantity'], autopct='%1.1f%%', colors=sns.color_palette('colorblind'))
axes[2].set_title('Quantity-wise Discount')

plt.tight_layout()
plt.show()
```

This code calculates the total sales, profit, and discount for each quantity of products. It visualizes the quantity-wise analysis using three bar plots.

### **Step 15: Category-wise Sales, Discount, and Profit Analysis**

```python
category_analysis = df.groupby('Category')[['Sales', 'Profit', 'Discount']].sum().reset_index()

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot category-wise sales
sns.barplot(ax=axes[0], data=category_analysis, x='Category', y='Sales', palette='pastel')
axes[0].set_title('Category-wise Sales')
axes[0].set_xlabel('Category')
axes[0].set_ylabel('Total Sales')

# Plot category-wise profit
sns.barplot(ax=axes[1], data=category_analysis, x='Category', y='Profit', palette='Set1')
axes[1].set_title('Category-wise Profit')
axes[1].set_xlabel('Category')
axes[1].set_ylabel('Total Profit')

# Plot category-wise discount
sns.barplot(ax=axes[2], data=category_analysis, x='Category', y='Discount', palette='Set2')
axes[2].set_title('Category-wise Discount')
axes[2].set_xlabel('Category')
axes[2].set_ylabel('Total Discount')

plt.tight_layout()
plt.show()
```

This code calculates the total sales, profit, and discount for each category of products. It visualizes the category-wise analysis using three bar plots.

### **Step 16: Sub-Category-wise Sales, Profit, and Discount Analysis**

```python
sub_category_analysis = df.groupby('Sub-Category')[['Sales', 'Profit', 'Discount']].sum().reset_index()

fig, axes = plt.subplots(1, 3, figsize=(15, 8))

# Plot sub-category-wise sales
sns.barplot(ax=axes[0], data=sub_category_analysis, x='Sub-Category', y='Sales', palette='muted')
axes[0].set_title('Sub-Category-wise Sales')
axes[0].set_xlabel('Sub-Category')
axes[0].set_ylabel('Total Sales')
axes[0].tick_params(axis='x', rotation=90)

# Plot sub-category-wise profit
sns.barplot(ax=axes[1], data=sub_category_analysis, x='Sub-Category', y='Profit', palette='Set3')
axes[1].set_title('Sub-Category-wise Profit')
axes[1].set_xlabel('Sub-Category')
axes[1].set_ylabel('Total Profit')
axes[1].tick_params(axis='x', rotation=90)

# Plot sub-category-wise discount
sns.barplot(ax=axes[2], data=sub_category_analysis, x='Sub-Category', y='Discount', palette='dark')
axes[2].set_title('Sub-Category-wise Discount')
axes[2].set_xlabel('Sub-Category')
axes[2].set_ylabel('Total Discount')
axes[2].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

This code calculates the total sales, profit, and discount for each sub-category of products. It visualizes the sub-category-wise analysis using three bar plots.

### **Step 17: Region-wise Analysis**

```python
region_analysis = df.groupby('Region')[['Sales', 'Profit', 'Discount']].sum().reset_index()

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot region-wise sales
sns.barplot(ax=axes[0], data=region_analysis, x='Region', y='Sales', palette='Set3')
axes[0].set_title('Region-wise Sales')
axes[0].set_xlabel('Region')
axes[0].set_ylabel('Total Sales')

# Plot region-wise profit
sns.barplot(ax=axes[1], data=region_analysis, x='Region', y='Profit', palette='Reds_r')
axes[1].set_title('Region-wise Profit')
axes[1].set_xlabel('Region')
axes[1].set_ylabel('Total Profit')

# Plot region-wise discount
sns.barplot(ax=axes[2], data=region_analysis, x='Region', y='Discount', palette='colorblind')
axes[2].set_title('Region-wise Discount')
axes[2].set_xlabel('Region')
axes[2].set_ylabel('Total Discount')

plt.tight_layout()
plt.show()
```

This code calculates the total sales, profit, and discount for each region. It visualizes the region-wise analysis using three bar plots.

### **Step 18: Ship Mode-wise Analysis**

```python
ship_mode_analysis = df.groupby('Ship Mode')[['Sales', 'Profit', 'Discount']].sum().reset_index()

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot ship mode-wise sales
sns.barplot(ax=axes[0], data=ship_mode_analysis, x='Ship Mode', y='Sales', palette='Blues_r')
axes[0].set_title('Ship Mode-wise Sales')
axes[0].set_xlabel('Ship Mode')
axes[0].set_ylabel('Total Sales')

# Plot ship mode-wise profit
sns.barplot(ax=axes[1], data=ship_mode_analysis, x='Ship Mode', y='Profit', palette='deep')
axes[1].set_title('Ship Mode-wise Profit')
axes[1].set_xlabel('Ship Mode')
axes[1].set_ylabel('Total Profit')

# Plot ship mode-wise discount
sns.barplot(ax=axes[2], data=ship_mode_analysis, x='Ship Mode', y='Discount', palette='muted')
axes[2].set_title('Ship Mode-wise Discount')
axes[2].set_xlabel('Ship Mode')
axes[2].set_ylabel('Total Discount')

plt.tight_layout()
plt.show()
```

This code calculates the total sales, profit, and discount for each ship mode. It visualizes the ship mode-wise analysis using three bar plots.
